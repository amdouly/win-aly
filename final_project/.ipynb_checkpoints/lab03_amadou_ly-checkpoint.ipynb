On remarque ici  les coorelation entre toutes les variables 

```python
sns.boxplot(y = "fixed acidity", data = dfr)
```

* fixed acidity (acidité fixer): la plupart des acides impliqués dans la vin ou fixes ou non volatile (ne s'evaporent pas facilement )
    * max 
    * min
    * moyenne

```python
sns.boxplot(y = "volatile acidity", data = dfr)
```

* volatile acidity(acidité volatile): la quantité d'acide acetique dans la vin qui a des niveaux trop elevé peut conduire a un gout de vinaigre desagreable
    * max
    * min 
    * moyenne 

```python
sns.boxplot(y = "citric acid", data = dfr)
```

* citric acid(acide citrique): trouvé en petites quantités ,l'acide citrique pet ajouter de la "fraicheur" et de la saveur aux vins
    * min
    * max 
    * moyenne

```python
sns.boxplot(y = "residual sugar", data = dfr)
```

* residual sugar(sucre résiduel): la quantité de sucre restant apres l'arret de la fermentation ,il est rare de trouver des vins avec 1g/L et les vins avec plus de 45g/L sont considere comme doux
    * min
    * max
    * moyenne 

```python
sns.boxplot(y = "chlorides", data = dfr)
```

* chlorides(chlorures):la quantité de sel dans le vin 
    * max
    * min
    * moyenne 

```python
sns.boxplot(y = "free sulfur dioxide", data = dfr)
```

* free sulfur dioxide (dioxyde de soufre libre):la forme libre de SO2 existe enequilibre entre le SO2 moleculaire (sous forme de gaz dissous) et l'ion bisulfate ; il empeche la croissance microbienne et l'oxydation du vin .
    * max 
    * min 
    * moyenne
