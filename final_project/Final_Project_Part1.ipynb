```python
def snake_case_col(name):
    """
    Snakecaseify a string
    :param name: str
        column name
    :return:
    """
    return name.strip().lower().replace(" ", "_")
```

```python
dft=dft.rename(columns=snake_case_col)
dft.head()
```

```python
# Normalize feature variable
#from sklearn.preprocessing import StandardScaler
#X_features = X
#X = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
strata = dft.color
dft_train, dft_test = train_test_split(dft, test_size=0.2, random_state=28, stratify=strata)
```
